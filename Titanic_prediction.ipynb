```python
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
df.describe()
```

```python
df.corr().round(1)
```

```python
features = ['Fare']
X = df[features]
Y = df['Survived']
```

```python
X.size
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
knn_model1 = KNeighborsClassifier()
param_grid = {'n_neighbors': np.arange(2,10)}
knn_gscv = GridSearchCV(knn_model1, param_grid, cv =5)
knn_gscv.fit(X,Y)
```

```python
knn_gscv.best_params_
```

```python
titanic_model = KNeighborsClassifier(n_neighbors = 8)
titanic_model.fit(X,Y)
y_pred = titanic_model.predict(X)
```

```python
y_pred
```

```python
titanic_model.score(X,Y)
```

```python
df
```

```python
df['Survival_Status'] = y_pred
```

```python
df['Survival_Status'].replace(0, 'NO', inplace = True)
```

```python
df['Survival_Status'].replace(1, 'YES', inplace = True)
```
